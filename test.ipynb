```python
# prompt: generate a list of 1 to 10 number

numbers = list(range(1, 11))
print(numbers)
array1 = numbers
#print(array1)
array2 = [x - 1 for x in array1]
#print(array2)
array3 = [ y-1 for y in array2]
#print(array3)
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot3D(array1,array2,array3)
plt.show()
```
